```python
import itertools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make five randomc countries
countries = ["A", "B", "C", "D", "E"]

# Years 2000 to 2030
years = range(2000, 2031)

# Make a dataframe
df = pd.DataFrame(list(itertools.product(countries, years)), columns=["Country", "Year"])
df = df.sort_values(by=["Country", "Year"])
df
```

```python
## STATS ##
df["GDP_Per_Capita"] = np.nan
df["Population"] = np.nan
df["Openness"] = np.nan

# Random starting population in thousands
start_pop = [3_000, 5_000, 2_000, 1_000, 4_000]
pop_growth_rate = [0.02, 0.03, 0.01, 0.05, 0.04]

for c in range(len(countries)):
    df.loc[df["Country"] == countries[c], "Population"] = start_pop[c] * (1 + pop_growth_rate[c]) ** (df["Year"] - 2000)


# random gdp per capita, larger differences
gdp_per_capita = [20_000, 30_000, 15_000, 10_000, 25_000]
gdp_growth_rate = [0.02, 0.03, 0.01, 0.05, 0.04]

for c in range(len(countries)):
    df.loc[df["Country"] == countries[c], "GDP_Per_Capita"] = gdp_per_capita[c] * (1 + gdp_growth_rate[c]) ** (df["Year"] - 2000)

df["GDP"] = df["GDP_Per_Capita"] * df["Population"]

# random openness to trade (0,1)
openness = [0.5, 0.6, 0.4, 0.3, 0.7]

for c in range(len(countries)):
    df.loc[df["Country"] == countries[c], "Openness"] = openness[c]

df
```
