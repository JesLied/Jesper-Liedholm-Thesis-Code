```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import util
```

```python
filename = "Population_By_Age_Group"
filepath = os.path.join("../Raw", filename)

df = pd.read_csv(filepath)
df
```

```python
# Only countries that exist in util.COUNTRIES class
df = df[df["geo"].isin(list(vars(util.COUNTRIES).values()))]
df
```

```python
# Parse the age group
age_group_dict = {
    'Proportion of population aged 0-14 years': [0, 14],
    'Proportion of population aged 15-24 years': [15, 24],
    'Proportion of population aged 25-49 years': [25, 49],
    'Proportion of population aged 50-64 years': [50, 64],
    'Proportion of population aged 65-79 years': [65, 79],
    'Proportion of population aged 80 years and more': [80, 100]
}

# Map against the age_group_dict and return the list
df["age_group_list"] = df["indic_de"].map(age_group_dict)
df[["Age_Group_Min_Age", "Age_Group_Max_Age"]] = pd.DataFrame(df["age_group_list"].tolist(), index=df.index)
df = df.drop(columns=["age_group_list"])
df
```

```python
# Now we want to save the cleaned data
# OBS_VALUE: Proportion of population
# TIME_PERIOD: Year
# age_group_start: Min age
# age_group_end: Max age
# geo: Country

df = df.rename(columns={
    "TIME_PERIOD": "Year",
    "OBS_VALUE": "Proportion_of_Population",
    "geo": "Country"
})
df = df[["Country", "Year", "Age_Group_Min_Age", "Age_Group_Max_Age", "Proportion_of_Population"]]
df
```

```python
# Save to Clean folder
clean_folder = "/Users/jesper/Desktop/CBS/Thesis 1/Data/Clean"
clean_filepath = os.path.join(clean_folder, filename + ".csv")
df.to_csv(clean_filepath, index=False)
```
