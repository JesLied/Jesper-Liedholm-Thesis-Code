```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import util
```

```python
main_df = pd.DataFrame(columns=["Country", "Year", "Metric", "Classification", "Value"])
main_df.shape
```

```python
# POPULATION
pop_df = pd.read_csv("/Users/jesper/Desktop/CBS/Thesis 1/Data/Clean/Population.csv")
pop_df["Value"] = pop_df["Population"]
pop_df["Metric"] = "Population"
pop_df["Classification"] = "Total"

pop_df = pop_df[['Country', 'Year', 'Metric', 'Classification', 'Value']]

main_df = pd.concat([main_df, pop_df], ignore_index=True)
main_df.shape
```

```python
# POPULATION BY AGE GROUP
pop_age_df = pd.read_csv("/Users/jesper/Desktop/CBS/Thesis 1/Data/Clean/Population_By_Age_Group.csv")
pop_age_df["Age_Group"] = pop_age_df["Age_Group_Min_Age"].astype(str) + "-" + pop_age_df["Age_Group_Max_Age"].astype(str)

pop_age_df["Metric"] = "Proportion_of_Population_By_Age_Group"
pop_age_df["Classification"] = pop_age_df["Age_Group"]
pop_age_df["Value"] = pop_age_df["Proportion_of_Population"]

pop_age_df = pop_age_df[['Country', 'Year', 'Metric', 'Classification', 'Value']]

main_df = pd.concat([main_df, pop_age_df], ignore_index=True)
main_df.shape
```

```python
# GDP
gdp_df = pd.read_csv("/Users/jesper/Desktop/CBS/Thesis 1/Data/Clean/GDP.csv")
gdp_df["Metric"] = "GDP"
gdp_df["Classification"] = "Total"
gdp_df["Value"] = gdp_df["GDP"]

gdp_df = gdp_df[['Country', 'Year', 'Metric', 'Classification', 'Value']]

main_df = pd.concat([main_df, gdp_df], ignore_index=True)
main_df.shape
```

```python
# PATENTS
pat_df = pd.read_csv("/Users/jesper/Desktop/CBS/Thesis 1/Data/Clean/Patents.csv")
pat_df["Metric"] = "Patents"
pat_df["Classification"] = "Total"
pat_df["Value"] = pat_df["Patents"]

pat_df = pat_df[['Country', 'Year', 'Metric', 'Classification', 'Value']]

main_df = pd.concat([main_df, pat_df], ignore_index=True)
main_df.shape
```

```python
# R&D GDP
rd_gdp_df = pd.read_csv("/Users/jesper/Desktop/CBS/Thesis 1/Data/Clean/R&D_GDP.csv")
rd_gdp_df["Metric"] = "R&D_GDP"
rd_gdp_df["Classification"] = rd_gdp_df["Sector"]
rd_gdp_df["Value"] = rd_gdp_df["R&D_GDP"]

rd_gdp_df = rd_gdp_df[['Country', 'Year', 'Metric', 'Classification', 'Value']]

main_df = pd.concat([main_df, rd_gdp_df], ignore_index=True)
main_df.shape
```
